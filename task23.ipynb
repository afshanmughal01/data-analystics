```python
# heart_eda.py / heart_eda.ipynb
# Complete EDA for UCI Cleveland Heart Disease dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import LabelEncoder
from itertools import combinations

# Optional: configure display
pd.set_option('display.max_columns', 200)
sns.set_style('whitegrid')

# 1) Load the dataset
# Replace the path with your local CSV path, or the raw link if accessible
#DATA_PATH = 'processed.cleveland.data.csv'  # <- change to your file name or path
df = pd.read_csv("heart.csv")

# 2) Inspect the data
print("Shape:", df.shape)
display(df.head())
print("\nInfo:")
print(df.info())
print("\nDescribe (all):")
display(df.describe(include='all'))

# 3) Clean data
# 3a. Standardize column names (many mirrors already have readable names; if not, rename)
# Example column names expected (if different in your CSV, adapt accordingly):
expected_cols = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','target']
if len(df.columns) == 14:
    df.columns = expected_cols

# 3b. Convert dtypes if necessary
# 'ca' and 'thal' sometimes have '?' strings in UCI raw; replace and convert
df.replace('?', np.nan, inplace=True)

# Convert numeric-looking columns
num_cols = ['age','trestbps','chol','thalach','oldpeak','ca']
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')

# target: in Cleveland, target values 0 = no disease, 1-4 = disease (often converted to binary)
if 'target' in df.columns:
    df['target_raw'] = df['target']
    # make binary target
    df['target'] = (df['target'].astype(float) > 0).astype(int)

# 3c. Missing values & duplicates
print("\nMissing values per column:")
print(df.isnull().sum())

# approach: drop rows with many missing values; impute numeric with median; categorical with mode
# Drop duplicates
df = df.drop_duplicates()
# Impute numeric medians
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
for c in numeric_cols:
    df[c].fillna(df[c].median(), inplace=True)
# Impute categorical (object/integers used as category)
cat_cols = [c for c in df.columns if df[c].dtype == 'object' or c in ['cp','restecg','slope','thal','ca','fbs','sex','exang']]
for c in cat_cols:
    if c in df.columns:
        df[c].fillna(df[c].mode().iloc[0], inplace=True)

print("\nAfter imputation missing counts:")
print(df.isnull().sum())

# 4) Univariate analysis
import matplotlib.ticker as ticker
print("\n--- Univariate: Numeric distributions ---")
num_cols = df.select_dtypes(include=[np.number]).columns.drop('target', errors='ignore')
for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')
    plt.tight_layout()
    plt.show()

print("\n--- Univariate: Categorical counts ---")
cat_cols = ['sex','cp','fbs','restecg','exang','slope','thal','ca']
for col in cat_cols:
    if col in df.columns:
        plt.figure(figsize=(6,3))
        sns.countplot(x=col, data=df)
        plt.title(f'Countplot of {col}')
        plt.tight_layout()
        plt.show()

# 5) Bivariate analysis
print("\n--- Bivariate numeric vs numeric scatterplots + correlation ---")
# a few key scatterplots
pairs = [('age','thalach'), ('chol','trestbps'), ('oldpeak','thalach')]
for x,y in pairs:
    if x in df.columns and y in df.columns:
        plt.figure(figsize=(5,4))
        sns.scatterplot(x=x, y=y, hue='target', data=df, alpha=0.8)
        plt.title(f'{y} vs {x} by target')
        plt.tight_layout()
        plt.show()

# Boxplots of numeric by target
for col in ['age','trestbps','chol','thalach','oldpeak']:
    if col in df.columns:
        plt.figure(figsize=(6,3))
        sns.boxplot(x='target', y=col, data=df)
        plt.title(f'{col} by target (0=no disease,1=disease)')
        plt.tight_layout()
        plt.show()

# Categorical vs target (bar plots)
for col in ['cp','sex','fbs','restecg','exang','slope','thal','ca']:
    if col in df.columns:
        plt.figure(figsize=(6,3))
        prop = df.groupby([col,'target']).size().unstack(fill_value=0)
        prop = prop.div(prop.sum(axis=1), axis=0)  # row proportions
        prop.plot(kind='bar', stacked=True, legend=True)
        plt.title(f'Proportion of target by {col}')
        plt.xlabel(col)
        plt.ylabel('Proportion')
        plt.tight_layout()
        plt.show()

# 6) Correlation analysis
print("\n--- Correlation matrix (numeric) ---")
numeric_cols = df.select_dtypes(include=[np.number]).columns
corr = df[numeric_cols].corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='vlag', center=0)
plt.title('Correlation matrix')
plt.tight_layout()
plt.show()

# For numeric vs binary target: compute point-biserial via Pearson (equivalent here)
print("\nNumeric correlation with target (absolute values):")
corr_with_target = corr['target'].drop('target', errors='ignore').sort_values(key=lambda x: x.abs(), ascending=False)
print(corr_with_target)

# For categorical vs numeric: compute group means and do ANOVA or Kruskal if non-normal
from scipy.stats import f_oneway, kruskal
print("\nCategorical vs numeric signal (example: 'cp' effect on thalach):")
if 'cp' in df.columns and 'thalach' in df.columns:
    groups = [group['thalach'].values for name, group in df.groupby('cp')]
    try:
        fstat, pval = f_oneway(*groups)
        print(f'ANOVA cp vs thalach: F={fstat:.3f}, p={pval:.3e}')
    except Exception as e:
        print("ANOVA error:", e)

# 7) Data drilling (aggregate then drill)
print("\n--- Data drilling example: overall -> sex -> cp -> records ---")
print("Overall counts by target:")
print(df['target'].value_counts())
print("\nBy sex and chest pain type (cp):")
grouped = df.groupby(['sex','cp'])['target'].agg(['count','mean','sum']).reset_index()
display(grouped.sort_values('count', ascending=False).head(20))

# drill into a small subgroup
sub = df[(df['sex']==1) & (df['cp']==3)]
print(f"\nSample records for sex=1, cp=3 (n={sub.shape[0]}):")
display(sub.head())

# 8) Outlier & anomaly detection
print("\n--- Outlier detection (IQR) ---")
def detect_outliers_iqr(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5*iqr
    upper = q3 + 1.5*iqr
    return series[(series < lower) | (series > upper)]

outlier_summary = {}
for col in ['age','trestbps','chol','thalach','oldpeak']:
    if col in df.columns:
        out = detect_outliers_iqr(df[col].dropna())
        outlier_summary[col] = out.shape[0]
print(outlier_summary)

# Z-score anomalies
num_for_z = df.select_dtypes(include=[np.number]).fillna(0)
z_scores = np.abs(stats.zscore(num_for_z))
anomaly_rows = (z_scores > 3).any(axis=1)
print("Anomaly rows by z>3:", anomaly_rows.sum())
if anomaly_rows.sum() > 0:
    display(df[anomaly_rows].head())

# 9) Pattern & trend discovery
print("\n--- Pattern discovery: simple clustering hint via KMeans (optional) ---")
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = df[numeric_cols.drop('target')].fillna(0)
X_scaled = sc.fit_transform(X)
kmeans = KMeans(n_clusters=3, random_state=42).fit(X_scaled)
df['kmeans3'] = kmeans.labels_
print("Cluster counts:")
print(df['kmeans3'].value_counts())
plt.figure(figsize=(6,4))
sns.countplot(x='kmeans3', hue='target', data=df)
plt.title('Clusters vs target')
plt.tight_layout()
plt.show()

# Frequent pattern: most common value combos for high-risk subgroup
print("\nTop frequent value combinations among patients with disease (target=1):")
disease = df[df['target']==1]
top_combos = disease.groupby(['cp','thal','slope','exang'])['age'].count().sort_values(ascending=False).head(10)
print(top_combos)

# 10) Interpretation (4-6 sentences)
print("\n--- Interpretation / Conclusion ---")
print("""
Interpretation:
1) Several numeric features (e.g., thalach - max heart rate, oldpeak - ST depression) show notable correlation with target; thalach tends to be lower for diseased patients.
2) Categorical features such as chest pain type (cp) and exercise-induced angina (exang) show strong differences in disease prevalence (certain cp categories have higher disease proportion).
3) A small number of outliers exist (e.g., extremely high cholesterol or very low thalach)  flag these for clinical review rather than automatic removal.
4) Clustering suggests groups of patients with different risk profiles; final action: use these descriptive insights to guide feature selection for predictive modeling and clinical investigation.
""")
```

```python
# ============================================================
# FULL EDA SCRIPT  Wine Quality Dataset
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# ------------------------------------------------------------
# 1. LOAD DATA
# ------------------------------------------------------------
df = pd.read_csv("WineQuality.csv")

# ------------------------------------------------------------
# 2. INSPECT DATA
# ------------------------------------------------------------
print("\n=== SHAPE ===")
print(df.shape)

print("\n=== DATA TYPES ===")
print(df.dtypes)

print("\n=== FIRST 5 ROWS ===")
print(df.head())

print("\n=== SUMMARY STATISTICS ===")
print(df.describe(include='all'))

print("\n=== MISSING VALUES ===")
print(df.isnull().sum())

# ------------------------------------------------------------
# 3. CLEANING
# ------------------------------------------------------------

# Remove duplicates
df = df.drop_duplicates()

# Impute numeric missing values using median
df = df.fillna(df.median(numeric_only=True))

# Convert quality to int (wine datasets usually have integer quality ratings)
if "quality" in df.columns:
    df["quality"] = df["quality"].astype(int)

print("\n=== After Cleaning  Missing Values ===")
print(df.isnull().sum())

# ------------------------------------------------------------
# 4. UNIVARIATE ANALYSIS
# ------------------------------------------------------------
numeric_cols = df.select_dtypes(include=["number"]).columns

print("\n=== UNIVARIATE HISTOGRAMS ===")
for col in numeric_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution: {col}")
    plt.show()

# Categorical example (quality)
if "quality" in df.columns:
    print(df["quality"].value_counts().sort_index())
    sns.countplot(x=df["quality"])
    plt.title("Wine Quality Distribution")
    plt.show()

# ------------------------------------------------------------
# 5. BIVARIATE ANALYSIS
# ------------------------------------------------------------

# Example: scatterplots vs. quality-related features
if "quality" in df.columns:
    for col in numeric_cols:
        if col != "quality":
            plt.figure(figsize=(6,3))
            sns.boxplot(x="quality", y=col, data=df)
            plt.title(f"{col} by Wine Quality")
            plt.show()

# ------------------------------------------------------------
# 6. CORRELATION ANALYSIS
# ------------------------------------------------------------
corr = df[numeric_cols].corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

if "quality" in df.columns:
    print("\n=== Top Correlations with Quality ===")
    print(corr["quality"].sort_values(ascending=False))

# ------------------------------------------------------------
# 7. DATA DRILLING (GRANULARITY)
# ------------------------------------------------------------

print("\n=== Drill-down: Overall Summary ===")
print(df.mean(numeric_only=True))

# Quality level summaries
if "quality" in df.columns:
    numeric_cols = df.select_dtypes(include=["number"]).columns
    df.groupby("quality")[numeric_cols].agg(["mean", "median", "count"])

    # Alcohol binning
    if "alcohol" in df.columns:
        df["alcohol_bin"] = pd.cut(df["alcohol"], bins=3, labels=["low","mid","high"])
        print("\n=== Drill-down: Quality  Alcohol Bin ===")
        print(df.groupby(["quality","alcohol_bin"]).size())

# ------------------------------------------------------------
# 8. OUTLIER & ANOMALY DETECTION
# ------------------------------------------------------------
print("\n=== IQR OUTLIER DETECTION ===")

def detect_iqr(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
    outliers = series[(series < lower) | (series > upper)]
    return outliers

for col in numeric_cols:
    out = detect_iqr(df[col])
    print(f"{col}: {len(out)} outliers")

# Z-score anomalies
z_scores = np.abs(stats.zscore(df[numeric_cols]))
anomalies = (z_scores > 3).any(axis=1)

print("\n=== Z-SCORE ANOMALIES ===")
print(f"Rows with anomalies: {anomalies.sum()}")

# ------------------------------------------------------------
# 9. PATTERN & TREND DISCOVERY
# ------------------------------------------------------------

# Example: k-means clustering
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

X = df[numeric_cols]
scaled = StandardScaler().fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42).fit(scaled)
df["cluster"] = kmeans.labels_

print("\n=== Cluster Sizes ===")
print(df["cluster"].value_counts())

print("\n=== Relation Between Cluster & Quality ===")
if "quality" in df.columns:
    print(pd.crosstab(df["quality"], df["cluster"]))

# ------------------------------------------------------------
# 10. INTERPRETATION
# ------------------------------------------------------------

print("\n=== INTERPRETATION (Summary) ===")
print(
    "The dataset shows clear relationships between wine quality and physicochemical properties. "
    "Alcohol has a strong positive association with quality, while volatile acidity is negatively associated. "
    "Outliers primarily appear in residual sugar and chlorides; these should be reviewed before modeling. "
    "Clustering reveals natural groupings in the wines, partially aligning with quality ratings. "
    "Overall, the dataset is clean and suitable for supervised ML prediction tasks such as regression or classification."
)
```
